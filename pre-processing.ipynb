```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import json
from tqdm import tqdm_notebook
```

## Preprocessing Uber Eats data

```python
df = pd.read_csv("./data/UE-NY-rest-details.csv")
df = df[['name', 'address']]
df['address'][101]
```

```python
df['address'].fillna("{}", inplace=True)
df['postalCode'] = None
df['addressLocality'] = None
df['addressRegion'] = None
df['streetAddress'] = None
```

```python
counter = 0
for i in tqdm_notebook(range(5), desc="preprocessing"):
    address = df['address'][i].replace("\'", "\"")
    try:
        obj = json.loads(address)
        df.at[i, 'postalCode'] = obj.get('postalCode')
        df.at[i, 'addressLocality'] = obj.get('addressLocality')
        df.at[i, 'addressRegion'] = obj.get('addressRegion')
        df.at[i, 'streetAddress'] = obj.get('streetAddress')
    except Exception as e:
        print(e)
        df.at[i, 'postalCode']  = None
        df.at[i, 'addressLocality'] = None
        df.at[i, 'addressRegion'] = None
        df.at[i, 'streetAddress'] = None
```
